```python
cursor.execute("DROP TABLE IF EXISTS location_dims;") #delete the table if it already exists

#create table with attributes based on the dimensional model
createTable1 = """CREATE TABLE location_dims(
                 fips int not null,
                 State VARCHAR(255),
                 State_Initial VARCHAR(255),
                 Country VARCHAR(255),
                 PRIMARY KEY (fips));"""

cursor.execute(createTable1)
db.commit() #commiting the create table statement

insertData1 = "INSERT INTO location_dims VALUES (%s,%s,%s,%s);" #insert data into table
cursor.executemany(insertData1,location_fips_list)
db.commit()
```

```python
cursor.execute("DROP TABLE IF EXISTS date_dims;") #delete the table if it already exists

#create table with attributes based on the dimensional model
createTable2 = """CREATE TABLE date_dims(
                 Date_Key VARCHAR(255),
                 Weekday INT,
                 Day_of_Week VARCHAR(255),
                 Week_Year INT,
                 Month INT,
                 Month_String VARCHAR(255),
                 Quarter INT,
                 Year INT,
                 PRIMARY KEY (Date_Key));"""

cursor.execute(createTable2)
db.commit() #commiting the create table statement

insertData2 = "INSERT INTO date_dims VALUES (%s,%s,%s,%s,%s,%s,%s,%s);" #insert data into table
cursor.executemany(insertData2,date_df_list)
db.commit()
```

```python
cursor.execute("DROP TABLE IF EXISTS demograph_dims;") #delete the table if it already exists

#create table with attributes based on the dimensional model
createTable3 = """CREATE TABLE demograph_dims(
                 State VARCHAR(255),
                 Age_Percent_65_Older FLOAT,
                 American_Indian FLOAT,
                 Asian FLOAT,
                 Black FLOAT,
                 Hispanic_Latino FLOAT,
                 Native_Hawaiian_Pacific FLOAT,
                 White FLOAT,
                 Median_Houseold_Income FLOAT,
                 Income_below_Poverty_Level FLOAT,
                 Population INT,
                 Population_Smile FLOAT,
                 PRIMARY KEY (State));"""

cursor.execute(createTable3)
db.commit() #commiting the create table statement

insertData3 = "INSERT INTO demograph_dims VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s);" #insert data into table
cursor.executemany(insertData3,demograph_df_list)
db.commit()
```

```python
cursor.execute("DROP TABLE IF EXISTS fact_tables;") #delete the table if it already exists

# create table with attributes based on the dimensional model
createTable7 = """CREATE TABLE fact_tables(
                  id VARCHAR(255),
                  state VARCHAR(255),
                  date VARCHAR(255),
                  fips INT,
                  cases INT,
                  deaths INT,
                  daily_cases INT,
                  daily_deaths INT,
                  Daily_Vac_Per_Mill FLOAT,
                  Retail_and_rec_PctChg FLOAT,
                  Grocery_and_pharmacy_PctChg FLOAT,
                  Parks_PctChg FLOAT,
                  Workplace_PctChg FLOAT,
                  Transit_PctChg FLOAT,
                  Residential_PctChg FLOAT,
                  PRIMARY KEY (id),
                  FOREIGN KEY (fips) REFERENCES location_dims(fips),
                  FOREIGN KEY (Date) REFERENCES date_dims(Date_Key));"""

cursor.execute(createTable7)
db.commit() #commiting the create table statement

insertData7 = "INSERT INTO fact_tables VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s);" #insert data into table
cursor.executemany(insertData7, fact_dim_final_list)
db.commit()

db.close() #closing the connection to MySQL Workbench
```
